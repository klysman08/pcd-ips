## Gráfico de linhas

O gráfico de linhas é uma das técnicas de visualização de dados mais simples e mais amplamente utilizadas. Um gráfico de linhas apresenta a informação como uma série de pontos de dados ou marcadores ligados por linhas retas. É possível personalizar a forma, o tamanho, a cor e outros elementos estéticos das linhas e dos marcadores para uma melhor clareza visual.

Aqui está uma lista Python que mostra a produção de maçãs (toneladas por hectare) ao longo de seis anos num país imaginário chamado Kanto.

````python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
````

Podemos visualizar como é que a produção de maçãs evolui ao longo do tempo com um gráfico de linhas. Para desenhar um gráfico de linhas, podemos usar a função `plt.plot`.

````python
plt.plot(yield_apples)
````

Chamar a função `plt.plot` resulta no desenho do gráfico de linhas, como esperado. A função retorna também uma lista de gráficos desenhados `[<matplotlib.lines.Line2D at 0x7ff70aa20760>]`, exibido com o output. Podemos incluir um ponto e vírgula (`;`) no final da última instrução na célula para evitar mostrar o output da célula, exibindo assim apenas o gráfico.

````python
plt.plot(yield_apples);
````

Vamos melhorar este gráfico passo a passo para o tornar mais informativo e apelativo

### Customizar o eixo dos X

O eixo dos X do gráfico exibe neste momento os índices dos elementos da lista de 0 a 5. O gráfico seria mais informativo se pudéssemos exibir o ano para o qual estamos a exibir os dados. Podemos fazer isto passando dois argumentos ao `plt.plot`.

````python
years = [2010, 2011, 2012, 2013, 2014, 2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
````

````python
plt.plot(years, yield_apples)
````

### Títulos de eixo

Podemos adicionar títulos aos eixos para ilustrar o que é que cada eixo representa usando os métodos `plt.xlabel` e `plt.ylabel`.

````python
plt.plot(years, yield_apples)
plt.xlabel('Ano')
plt.ylabel('Produção (tons por hectare)');
````

### Traçar Múltiplas Linhas

Podemos invocar a função `plt.plot` uma vez para cada linha para traçar várias linhas no mesmo gráfico. Vamos comparar a produção de macãs vs. laranjas em Kanto.

````python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
````

````python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Ano')
plt.ylabel('Produção (tons por hectare)');
````

### Título do Gráfico e Legenda

Para diferenciar entre múltiplas linhas, podemos incluir uma legenda no gráfico usando a função `plt.legend`. Podemos também definir um título para o gráfico com a função `plt.title`.

````python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Ano')
plt.ylabel('Produção (tons por hectare)')

plt.title("Produção das Colheitas em Kanto")
plt.legend(['Maçãs', 'Laranjas']);
````

### Marcadores de Linha

Também é possível exibir marcadores para os pontos de dados em cada linha usando o argumento `marker` de `plt.plot`. O Matplotlib oferece diversos markers, como um círculo, cruz, quadrado, diamante, etc. Podemos ver a lista completa de tipos de marcadores aqui: https://matplotlib.org/3.1.1/api/markers_api.html .

````python
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')

plt.xlabel('Ano')
plt.ylabel('Produção (tons por hectare)')

plt.title("Produção das Colheitas em Kanto")
plt.legend(['Maçãs', 'Laranjas']);
````

### Estilizar Linhas e Marcadores

A função `plt.plot` suporta diversos argumentos para estilizar linhas e marcadores:

- `color` ou `c`: Definir a cor da linha ([cores suportadas](https://matplotlib.org/3.1.0/gallery/color/named_colors.html))
- `linestyle` ou `ls`: Escolher entre uma linha sólida ou tracejada
- `linewidth` ou `lw`: Definir a grossura da linha
- `markersize` ou `ms`: Definir o tamanho dos marcadores
- `markeredgecolor` ou `mec`: Definir a cor da linha para os marcadores
- `markeredgewidth` ou `mew`: Definir a grossura da linha para os marcadores
- `markerfacecolor` ou `mfc`: Definir a cor de fundo para os marcadores
- `alpha`: Opacidade do gráfico


Consulte a documentação de `plt.plot` para saber mais: [https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) .

````python
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Ano')
plt.ylabel('Produção (tons por hectare)')

plt.title("Produção das Colheitas em Kanto")
plt.legend(['Maçãs', 'Laranjas']);
````

O argumento `fmt` oferece uma forma abreviada de especificar o formato do marcador, e estilo e cor de linha. Pode ser fornecido como terceiro argumento a `plt.plot`.

```
fmt = '[marcador][linha][cor]'
```



````python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel('Ano')
plt.ylabel('Produção (tons por hectare)')

plt.title("Produção das Colheitas em Kanto")
plt.legend(['Maçãs', 'Laranjas']);
````

Se não especificarmos um estilo de linha em `fmt`, são desenhados apenas os marcadores.

````python
plt.plot(years, oranges, 'or')
plt.title("Produção de Laranjas (tons por hectare)");
````

## Gráficos de dispersão

Num gráfico de dispersão *(scatter plot)*, os valores de 2 variáveis são desenhados como pontos numa grelha de 2 dimensões. Adicionalmente, podemos usar uma terceira variável para determinar o tamanho ou cor dos pontos. Vamos tentar um exemplo.

O [Dataset Flores de Íris](https://en.wikipedia.org/wiki/Iris_flower_data_set) fornece amostras de medidas de sépalas e pétalas para três espécies de flores de íris. O dataset Íris vem incluído na biblioteca Seaborn e pode ser carregado como um dataframe Pandas.

````python
# Carregar os dados num dataframe Pandas
flowers_df = sns.load_dataset("iris")
````

````python
flowers_df
````

````python
flowers_df.species.unique()
````

Vamos tentar visualizar a relação entre o comprimento e a largura da sépala. O nosso primeiro instinto poderá ser criar um gráfico de linhas com `plt.plot`.

````python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
````

O resultado não é muito informativo dado que existem muitas combinações das duas propriedades no dataset. Não parece existir uma relação simples entre eles.

Podemos usar um gráfico de dispersão para visualizar como é que o comprimento e a largura das sépalas variam usando a função `scatterplot` do módulo `seaborn` (importado como `sns`).

````python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width);
````

### Adicionar Tonalidades de Cor

Repare como os pontos no gráfico acima parecem formar diferentes clusters com alguns outliers. Podemos dar cor aos pontos usando a espécie das flores para determinar a tonalidade ou matiz (`hue`). Podemos também aumentar o tamanho dos pontos com o argumento `s`.

````python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width, hue=flowers_df.species, s=100);
````

Adicionar as cores torna o gráfico mais informativo. Podemos perceber de imediato que as flores Setosa têm um comprimento de sépala menor mais uma largura de sépala maior. Em contrapartida, acontece o contrário com as flores Virginica. 

### Customizar Figuras Seaborn

Dado que o Seaborn usa internamente as funções de gráficos do Matplotlib, podemos usar funções como `plt.figure` e `plt.title` para modificar a figura.

````python
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimensions')

sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species,
                s=100);
````

### Desenho de Gráficos com Dataframes Pandas

O Seaborn tem suporte incorporado de dataframes Pandas. Em vez de passar cada coluna como uma série, podemos fornecer os nomes das colunas e usar o argumento `data` para especificar um dataframe.

````python
plt.title('Sepal Dimensions')
sns.scatterplot(x='sepal_length', 
                y='sepal_width', 
                hue='species',
                s=100,
                data=flowers_df);
````

## Histograma

Um histograma representa a distribuição de uma variável criando classes ou *bins* (intervalos) ao longo da gama de valores e exibindo barras verticais para indicar o número de observações em cada classe.

Por exemplo, vamos visualizar a distribuição de valores de largura das sépalas no dataset das flores. Podemos usar a função `plt.hist` para criar um histograma.

````python
# Carregar os dados para um dataframe Pandas
flowers_df = sns.load_dataset("iris")
````

````python
flowers_df.sepal_width
````

````python
plt.title("Distribuição da Largura das Sépalas")
plt.hist(flowers_df.sepal_width);
````

Podemos ver imediatamente que as larguras das sépalas situam-se na gama 2.0 - 4.5, e cerca de 35 valores estão na gama dos 3.0, que parece ser a classe mais populada.

### Controlar o tamanho e número de classes

Podemos controlar o número de classes ou o tamanho de cada classe *(bin)* com o argumento `bins`


````python
# Especificar o número de classes
plt.hist(flowers_df.sepal_width, bins=5);
````

````python
import numpy as np

# Especificar os limites de cada classe
plt.hist(flowers_df.sepal_width, bins=np.arange(2, 5, 0.25));
````

````python
# Classes de tamanhos diferentes
plt.hist(flowers_df.sepal_width, bins=[1, 3, 4, 4.5]);
````

### Histogramas Múltiplos

Tal como nos gráficos de linhas, podemos desenhar múltiplos histogramas num único gráfico. Podemos reduzir a opacidade de cada histograma para que as barras de um histograma não escondam as do outro.

Vamos desenhar histogramas separados para cada espécie de flor.

````python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
````

````python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
````

Também é possível empilhar múltiplos histogramas uns em cima dos outros.

````python
plt.title("Distribuição da Largura das Sépalas")

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
````

## Gráfico de Barras

Os gráficos de barras são muito semelhantes aos gráficos de linhas, i.e, mostram uma sequência de valores. Contudo, é mostrada uma barra para cada valor, em vez de pontos ligados por linhas. Podemos usar a função `plt.bar` para desenhar um gráfico de barras.

````python
years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
````

````python
plt.bar(years, oranges);
````

Tal como nos histogramas, podemos emplilhar as barras umas em cima das outras. Para tal, usamos o argumento `bottom` de `plt.bar`.

````python
plt.bar(years, apples)
plt.bar(years, oranges, bottom=apples);
````

### Gráficos de Barras com Médias

Vamos ver outro dataset exemplo incluído com o Seaborn, chamado `tips`. O dataset contém informação sobre o sexo, hora do dia, total da conta, e valor da gorjeta para os clientes que visitam um restaurante ao longo de uma semana.

````python
tips_df = sns.load_dataset("tips");
````

````python
tips_df
````

Podemos querer desenhar um gráfico de barras para visualizar como é que varia em média o valor da conta em diferentes dias da semana. Uma forma de fazer isto seria calcular as médias diárias e depois usar `plt.bar` (pode tentar fazer isso como exercício).

As linhas a cortar cada barra representam o nível de variação nos valores. Por exemplo, parece que a variação no valor da conta é relativamente elevado às sextas-feiras e baixo ao sábado.

````python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df);
````

Podemos tornar as barras horizontais trocando simplesmente os eixos

````python
sns.barplot(x='total_bill', y='day', hue='sex', data=tips_df);
````

## Heatmap

Os heatmaps são utilizados para visualizar dados bidimensionais como uma matriz ou tabela usando cores. A melhor forma de perceber é através de um exemplo. Vamos usar outro dataset de referência do Seaborn, chamado `flights`, para visualizar o tráfego mensal de passageiros num aeroporto ao longo de 12 anos.

````python
flights_df = sns.load_dataset("flights").pivot_table(index="month", columns="year", values="passengers")
````

````python
flights_df
````

`flights_df` é uma matriz com uma linha para cada mês e uma coluna para cada ano. Os valores mostram o número de passageiros (em milhares) que visitaram o aeroporto num mês específico de um ano. Podemos usar a função `sns.heatmap` para visualizar o fluxo de pessoas no aeroporto.

````python
plt.title("Nº de passageiros (1000s)")
sns.heatmap(flights_df);
````

As cores mais brilhantes denotam uma maior afluência de pessoas ao aeroporto. Ao olhar para o gráfico, podemos inferir duas coisas:

- A afluência de pessoas ao aeroporto num determinado ano tende a ser mais elevada por volta de julho e agosto.
- A afluência de pessoas ao aeroporto num determinado mês tende a aumentar de ano para ano.

Também podemos apresentar os valores reais em cada bloco especificando `annot=True` e utilizando o argumento `cmap` para alterar a paleta de cores.

````python
plt.title("Nº de passageiros (1000s)")
sns.heatmap(flights_df, fmt=".0f", annot=True, cmap='Blues');
````

## Desenhar múltiplos gráficos numa grelha

O Matplotlib e o Seaborn também suportam o desenho de múltiplos gráficos numa grelha, usando `plt.subplots`, que retorna um conjunto de eixos para o desenho de gráficos. 

Aqui está uma grelha única com os diferentes tipos de gráficos que cobrimos nesta aula.

````python
fig, axes = plt.subplots(2, 3, figsize=(16, 8))

# Usar os eixos para o desenho de gráficos
axes[0,0].plot(years, apples, 's-b')
axes[0,0].plot(years, oranges, 'o--r')
axes[0,0].set_xlabel('Ano')
axes[0,0].set_ylabel('Produção (tons por hectare)')
axes[0,0].legend(['Maçãs', 'Laranjas']);
axes[0,0].set_title('Produção das Colheitas em Kanto')


# Passar os eixos ao seaborn
axes[0,1].set_title('Comprimento das Sépalas vs. Largura das Sépalas')
sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species, 
                s=100, 
                ax=axes[0,1]);

# Usar os eixos para fazer gráficos
axes[0,2].set_title('Distribuição da Largura das Sépalas')
axes[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

axes[0,2].legend(['Setosa', 'Versicolor', 'Virginica']);

# Passar os eixos ao seaborn
axes[1,0].set_title('Contas de Restaurante')
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, ax=axes[1,0]);

# Passar os eixos ao seaborn
axes[1,1].set_title('Tráfego aéreo')
sns.heatmap(flights_df, cmap='Blues', ax=axes[1,1]);

# Desenhar uma imagem usando os eixos
axes[1,2].set_title('Meme de Data Science')
axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])

plt.tight_layout(pad=2);
````
